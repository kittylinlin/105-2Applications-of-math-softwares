```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
```

```python
def draw(n):
    e = n*np.random.randn(365)
    x = np.linspace(0,12,365)
    y = np.sin(2*x)/(1+x)
    plt.scatter(x,y+e,color='g',marker='^')
    plt.scatter(x,y,color='r',s=10, alpha=.3)
```

```python
interact(draw, n=(-0.5,0.5,0.001))
```
