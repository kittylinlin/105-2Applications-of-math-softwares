```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris.data #花萼長度、寬度、花瓣長度、寬度
```

```python
iris.target #0, 1, 2三種不同花
```

```python
x = iris.data[:, 2:] #花瓣長度、寬度
y = iris.target
```

```python
plt.scatter(x[:,0], x[:,1], c=y, s=50)
```

```python
from sklearn.svm import SVC
```

```python
clf = SVC()
```

```python
clf.fit(x,y)
```

```python
gd = np.array([(i, j) for i in np.arange(0, 8, 0.5) for j in np.arange(-0.5, 3, 0.25)])
```

```python
plt.scatter(gd[:,0], gd[:,1], s=50, c=clf.predict(gd))
```

## 換個顏色試試

```python
plt.scatter(x[:,0], x[:,1], c=y, s=50, cmap=plt.cm.cool, marker="p")
```
