# Step 1: 先把該讀入的套件讀一讀吧！

```python
from keras.models import Sequential
from keras.layers.core import Dense, Activation, Dropout
from keras.layers.recurrent import LSTM
import numpy as np
```

# Step 2: 讀入訓練資料
* 也就是一堆英文名字啦！

* 男生的英文名字，總共有2943筆資料

```python
with open("male.txt") as f: # 讀入male.txt
    m_names = f.readlines() # 逐行讀取
m_names
```

* 女生的英文名字，總共有5001筆資料

```python
with open("female.txt") as f: # 讀入female.txt
    f_names = f.readlines() # 逐行讀取
f_names
```

# Step 3: 將訓練資料整理一下吧！
* 有些名字可能男女通用，這樣就不符合函數的定義啦！所以要先將這些中性的名字去除
* 為了讓資料更整齊，將所有字母都小寫化，並且將特殊符號移除

```python
mf_names = [] # 記錄m_names和f_names的交集
for f_name in f_names:
    if f_name in m_names:
        mf_names.append(f_name)
```

```python
# 去除m_names中的f_names，並小寫化和移除特殊符號
m_names = [m_name.lower().strip().replace("'",'').replace('-','').replace(' ','') for m_name in m_names if not m_name in mf_names]

# 去除m_names中的f_names，並小寫化和移除特殊符號
f_names = [f_name.lower().strip().replace("'",'').replace('-','').replace(' ','') for f_name in f_names if not f_name in mf_names]
```

* 計算所需維度，並建立相應大小之零矩陣

```python
# 名字總筆數
totalEntries = len(m_names) + len(f_names)

# m_names最長的名字長度＋f_names最長的名字長度
maxlen = len(max(m_names, key=len)) + len(max(f_names, key=len))

# 將m_names和f_names中的名字用""連(join)起來後，將字串的字元建立集合物件(同字母收集成一個)
chars = set("".join(m_names) + "".join(f_names))

# 將集合chars建立dic
char_indices = dict((element, index) for index, element in enumerate(chars))
```

```python
# x為input name的矩陣(第幾筆資料-名字中的字母位置k-名字中的第k個字母)
x = np.zeros((totalEntries, maxlen, len(chars)))

# y為output gender的矩陣(第幾筆資料-男生:1/女生:0)
y = np.zeros((totalEntries, 1))
```

* 將訓練資料逐筆放入矩陣中

```python
# 將m_names一一對應入x和y
for index, name in enumerate(m_names):
    for t, char in enumerate(name):
        x[index, t, char_indices[char]] = 1
    y[index] = 1
```

```python
# 將f_names一一對應入x和y
for index, name in enumerate(f_names):
    for t, char in enumerate(name):
        x[index + len(m_names), t, char_indices[char]] = 1
    y[index + len(m_names)] = 0
```

# Step 4: 建立model -> 訓練 -> 準確度

```python
model = Sequential()
model.add(LSTM(512, return_sequences=True, input_shape=(maxlen, len(chars))))
model.add(Dropout(0.2))
model.add(LSTM(512, return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(1))
model.add(Activation('sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(x, y, batch_size=16, epochs=10)
```

```python
loss, acc = model.evaluate(x, y, batch_size=16)
loss, acc
```

# Step 5: 開始來測試吧！

```python
from ipywidgets import interact_manual
```

```python
def predict(name):
    name_l = name.lower()
    x = np.zeros((1, maxlen, len(chars)))
    for t, char in enumerate(name_l):
        x[0, t, char_indices[char]] = 1
    preds = model.predict(x, verbose=0)
    if preds > 0.5:
        print("%s is male" %(name))
    else:
        print("%s is female" %(name))
```

```python
interact_manual(predict, name="Name");
```
