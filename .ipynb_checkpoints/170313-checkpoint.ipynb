# 機器學習概要

## SVM model

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([[-3, 2], [-6, 5], [0, -4], [2, -8]])
y = np.array([1, 2, 3, 4])
```

```python
x[:,0]
```

```python
plt.scatter(x[:,0], x[:,1], c = y, s=50)
```

```python
from sklearn.svm import SVC
```

```python
clf = SVC() #開一個分類器
```

```python
clf.fit(x,y) #把x,y放進SVM model
```

```python
clf.predict([[-3, 2]])
```

```python
clf.predict([[2, -8]])
```

```python
[i for i in range(5)]
```

```python
[(i, j) for i in range(4) for j in range(5)]
```

```python
np.arange(0, 10, 0.7)
```

```python
gd = np.array([(i, j) for i in np.arange(-8, 4, 0.6) for j in np.arange(-10, 6, 0.8)])
```

```python
gdc = clf.predict(gd)
```

```python
plt.scatter(gd[:,0], gd[:,1], s=50, c=gdc)
```

## K-Means

```python
x = np.random.rand(100, 2)
```

```python
plt.scatter(x[:,0], x[:,1], s=50)
```

```python
from sklearn.cluster import KMeans
```

```python
clf = KMeans(n_clusters=3) # 分三群
```

```python
clf.fit(x)
```

```python
clf.labels_
```

```python
plt.scatter(x[:,0], x[:,1], s=50, c=clf.labels_)
```

```python
gd = np.array([(i, j) for i in np.arange(-0.2, 1.2, 0.05) for j in np.arange(-0.2, 1.2, 0.05)])
```

```python
plt.scatter(gd[:,0], gd[:,1], s=50, c=clf.predict(gd))
```

## SVM

```python
from sklearn.datasets import make_classification
```

```python
x, y = make_classification(n_features=2, n_redundant=0, n_informative=2, n_clusters_per_class=1, n_classes=3)
```

```python
plt.scatter(x[:,0], x[:,1], s=50, c=y)
```

```python
clf = SVC()
```

```python
clf.fit(x,y)
```

```python
gd = np.array([(i, j) for i in np.arange(-4, 4, 0.5) for j in np.arange(-4, 4, 0.5)])
```

```python
plt.scatter(gd[:,0], gd[:,1], s=50, c=clf.predict(gd))
```
