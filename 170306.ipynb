## 測試線性資料
$$f(x) = 1.64x + 3x$$

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 100)
y = 1.64*x + 3 + 0.78*np.random.randn(100)
plt.scatter(x,y)
```

```python
def f(x, a, b):
    return a*x + b
```

* 利用資料和最小平方法，找出函數

```python
from scipy.optimize import curve_fit
```

```python
popt, pcov = curve_fit(f, x, y) #回傳最佳化後的參數, 最佳化後的共變異數
```

```python
a, b = popt
a, b
```

```python
plt.scatter(x, y)
plt.plot(x, a*x + b, 'r')
```

```python
#預測未來資料
def g(x):
    return a*x + b
```

```python
g(12)
```

* 機器學習的科學工具.交互驗證/自行切割資料

```python
from sklearn.cross_validation import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
len(x_train)
```

```python
popt, pcov = curve_fit(f, x_train, y_train) #利用訓練資料找出參數
a, b = popt
plt.scatter(x_test, y_test) #測試資料
plt.plot(x_test, a * x_test + b, 'r')
```

```python
#計算誤差
MSE_train = np.average(((a * x_train + b) - y_train)**2)
MSE_test = np.average(((a * x_test + b) - y_test)**2)
MSE_train, MSE_test
```

## 測試非線性資料
$$f(x) = sin(3.2x) + 1.8x$$

```python
y = np.sin(3.2*x) + 1.8*x + 0.38*np.random.randn(100)
```

```python
plt.scatter(x,y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
def f(x, a, b):
    return np.sin(a*x) + b*x
```

```python
popt, pcov = curve_fit(f, x_train, y_train) #利用訓練資料找出參數
a, b = popt
plt.scatter(x_train, y_train) #訓練資料
plt.plot(x, np.sin(a*x) + b*x, 'r')
```

```python
plt.scatter(x_train, y_train)
plt.plot(x_train, np.sin(a*x_train) + b*x_train, 'r') # x_train的排序會亂掉
```

```python
plt.scatter(x_train, y_train)
plt.plot(np.sort(x_train), np.sin(a*np.sort(x_train)) + b*np.sort(x_train), 'r') # 重新排序x_train
```

* 增加起始值讓訓練更準確

```python
popt, pcov = curve_fit(f, x_train, y_train, p0=[3,1.5]) #增加起始值
a, b = popt
plt.scatter(x_train, y_train) #訓練資料
plt.plot(x, np.sin(a*x) + b*x, 'r')
```
